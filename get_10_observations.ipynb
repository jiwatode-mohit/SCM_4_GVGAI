# First 10 observations

```python
### First 10 observations of each game ###

for j in range(5):
    for game in game_list:
        print(f"Playing {game}...")
        env = gym.make(f'gvgai-{game}-lvl0-v0')
        env.reset()
        root_folder_path = f"/home/jiwatode/projects/nobackup/gvgai/observations"
        
        # With game description
        folder_path_w_desc = root_folder_path + "/first_10/" + str(j)
        if not os.path.exists(folder_path_w_desc):
            os.makedirs(folder_path_w_desc, exist_ok=True)
        

        action_meanings_dict = {x: y for x, y in enumerate(env.unwrapped.get_action_meanings())}
        

        # Erase the text files before writing new info to it
        for folder_path in [
            folder_path_w_desc]:
            if os.path.exists(f"{folder_path}/{game}.txt"):
                open(f"{folder_path}/{game}.txt", "w").close()


        
        score = 0
        for i in range(10):

            action_id = env.action_space.sample()
            state, reward, isOver, info = env.step(action_id)
            ascii_grid_ = info["ascii"]
            max_len = max(len(cell) for r in ascii_grid_ for cell in r)
                
            with open(f"{folder_path_w_desc}/{game}.txt", "a") as f:
                ascii_grid = [row.split(',') for row in info["ascii"].split('\n')]
                f.write(f"Action: {action_meanings_dict[action_id]}\n")
                for row in ascii_grid:
                    f.write(' '.join(f"{cell:{max_len}}" for cell in row) + "\n")
                f.write("\n \n")

            if isOver or i == 9:
                print(f"Game {game} over at game tick " + str(i+1) + " with player " + info['winner'])
                break
```
